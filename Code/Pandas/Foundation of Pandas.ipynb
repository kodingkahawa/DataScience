`index_col` is used to select which of the columns/featues can be used as an index in the data:

```python
df = pd.read_csv('acs2015_census_tract_data.csv',\
                 low_memory=False, 
                 index_col = 'CensusTract')
```

```python
df.head()
```

`index_col` can also receive several indices of columns in a list:

```python
df = pd.read_csv('acs2015_census_tract_data.csv',\
                 low_memory=False, 
                 index_col = ['CensusTract', 'State', 'Men', 'Women'])
```
